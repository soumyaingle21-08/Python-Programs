```python
i = 1
while i <= 20:
    print("Hello Students")
    i = i + 1
```

```python
num = int(input("Enter a number: "))
sum = 0
i = 1

while i <= num:
    sum = sum + i
    i = i + 1

print("Sum =", sum)
```

```python
num = int(input("Enter a number: "))

for i in range(1, 11):
    print(num, "x", i, "=", num * i)
```

```python
num = 1

for i in range(1,5):
    for j in range(i):
        print(num, end=" ")
        num = num + 1
    print()
```

```python
for i in range(1,6):
    for j in range(i):
        print("*", end="")
    print()
```

```python
for i in range(1,6):
    for j in range(6-i):
        print(i, end=" ")
    print()
```

```python
for i in range(5,0,-1):
    for j in range(i,0,-1):
        print(j, end=" ")
    print()
```

```python
for i in range(1,6):
    for j in range(5-i):
        print(" ", end=" ")
    for k in range(1,i+1):
        print(k, end=" ")
    print()
```

```python
for i in range(1,6):
    for j in range(5-i):
        print(" ", end="")
    for k in range(i):
        print("* ", end="")
    print()
```
