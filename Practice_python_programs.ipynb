```python
str="SBJAIN college 56789"
dig=0
alpha=0
space=0
for i in str:
    if i.isdigit():
      dig+=1
    elif i.isalpha():
      alpha+=1
    else:
      space+=1
print(dig)
print(alpha)
print(space)
```
