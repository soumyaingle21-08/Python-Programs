```python
import numpy as np

arr = np.array([[1,2,3],[4,5,6]])

print("Array:\n", arr)

print("Number of Dimensions:", arr.ndim)
print("Shape of Array:", arr.shape)
print("Size of Array:", arr.size)
print("Type of Elements:", arr.dtype)
```

```python
count = [0,0,0,0,0]
spoil = 0

n = int(input("Enter number of voters: "))

for i in range(n):
    vote = int(input("Enter candidate number (1-5): "))

    if 1 <= vote <= 5:
        count[vote-1] += 1
    else:
        spoil += 1

for i in range(5):
    print("Votes for Candidate", i+1, ":", count[i])

print("Spoilt Ballots:", spoil)
```

```python
arr = []
n = int(input("Enter number of elements: "))

for i in range(n):
    num = int(input("Enter element: "))
    arr.append(num)

max_val = max(arr)
min_val = min(arr)

print("Maximum element:", max_val)
print("Minimum element:", min_val)
print("Maximum Difference:", max_val - min_val)
```

```python
import numpy as np

arr1 = np.array([0,10,20,40,60,80])
arr2 = np.array([10,30,40,50,70,90])

result = np.setdiff1d(arr1, arr2)

print("Array1:", arr1)
print("Array2:", arr2)
print("Set difference between two arrays:", result)
```

```python
arr = []

for i in range(10):
    num = int(input("Enter number: "))
    arr.append(num)

rev_arr = arr[::-1]

print("Original Array:", arr)
print("Reversed Array:", rev_arr)
```
