```python
s = input("Enter string: ")

upper = 0
lower = 0

for i in s:
    if i.isupper():
        upper += 1
    elif i.islower():
        lower += 1

print("Uppercase:", upper)
print("Lowercase:", lower)
```

```python
s = input("Enter string: ")

letters = 0
digits = 0

for i in s:
    if i.isalpha():
        letters += 1
    elif i.isdigit():
        digits += 1

print("Letters:", letters)
print("Digits:", digits)
```
